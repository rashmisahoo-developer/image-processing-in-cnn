```python
import PIL,google.colab
import matplotlib.pyplot as plt
from PIL import ImageEnhance
```

```python
google.colab.files.upload()
image=PIL.Image.open('img.jpg')
plt.imshow(image)
```

**Image Rotation**

```python
for i in range(0,180,10):
  rotate=image.rotate(i)
  plt.imshow(rotate)
```

**Blurred Image**

```python
image1=image.resize((50,50))
plt.imshow(image1)
```

```python
print(image.size)
```

Image Brightness



```python
sharpness=ImageEnhance.Sharpness(image)
sharpness_image=sharpness.enhance(11)
plt.imshow(sharpness_image)
```

**image contrast**

```python
contrast = ImageEnhance.Contrast(image)
constrast_image=contrast.enhance(6)
plt.imshow(constrast_image)
```

```python
color=ImageEnhance.Color(image)
color_image=color.enhance(8)
plt.imshow(color_image)
```
